```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import gensim.downloader
import numpy as np
```

```python
r = requests.session()

page = r.get(
    "https://www.nature.com/search?article_type=research&subject=climate-sciences, ecology, energy-science-and-technology, environmental-sciences, environmental-social-sciences, hydrology, natural-hazards, ocean-sciences, planetary-science, social-sciences, solid-earth-sciences, water-resources&order=relevance"
).content
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
links = []
date = []
headlines = []
```

```python
def toData(link):
    page = r.get(link).content
    soup = BeautifulSoup(page, "html.parser")
    for curr in soup.find_all(
        "div", {"class": "c-card__body u-display-flex u-flex-direction-column"}
    ):
        curr = str(curr)
        forHead = curr
        forHead = forHead[forHead.index(' itemprop="url">') + 16 :]
        forHead = forHead[: forHead.index("</a>")]
        headlines.append(forHead)
        url = curr
        url = url[url.index('href="/articles/') + 6 : url.index(' itemprop="url">') - 1]
        links.append("https://www.nature.com" + url)
    j = 0
    for curr in soup.find_all("time"):
        date.append(curr["datetime"])
        j += 1
```

```python
def checkIfSame():
    print(len(headlines))
    return len(headlines) == len(date) == len(links)
```

```python
allLink = [
    "https://www.nature.com/search?order=relevance&subject=climate-sciences&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=environmental-sciences&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=environmental-social-sciences&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=hydrology&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=energy-science-and-technology&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=natural-hazards&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=ocean-sciences&article_type=research&page=",
    "https://www.nature.com/search?order=relevance&subject=solid-earth-sciences&article_type=research&page=",
]
```

```python
for i in allLink:
    for j in range(20):
        toData(i + str(j + 1))
```

```python
checkIfSame()
```

```python
data = pd.DataFrame(
    list(zip(date, links, headlines)), columns=["date", "url", "headlines"]
)
```

```python
data = data.drop_duplicates(subset="headline")
```

```python
glove_vectors = gensim.downloader.load("glove-wiki-gigaword-50")
```

```python
def toVector(sentence):
    sentence = sentence.split(" ")
    currList = []
    for i in sentence:
        try:
            curr = glove_vectors[i.lower()]
            currList.append(curr)
        except:
            continue
    returnData = np.mean(currList, axis=0).tolist()
    if isinstance(returnData, list):
        return returnData
    return "delete"
```

```python
data["embedding"] = data.headline.apply(toVector)
```
