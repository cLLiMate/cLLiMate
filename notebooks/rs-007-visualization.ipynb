```python
import pandas as pd
import pyarrow
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_parquet("../data/processed/news-consolidated-v1.parquet")
data["years"] = data["date"].dt.year
data = data.sort_values(by="date")
data = data[40:]
nature = data[data["source"] == "nature"]
nature["years"].value_counts().sort_index().to_json("../data/vis/numNature.json", orient="split")
nature["years"].value_counts().sort_index().plot(kind="bar", title="Nature")
plt.show()
print("Insight: Nature has not been publishing more articles over the years. It has remained at roughly 10.5k throughout the 2010s. Yet, the spike indicates an exponential increase in climate papers. This proves that a greater chunk of submissions are regarding climate and environmental concerns.")
climate = data[data["source"] == "climate-db"]
climate["years"].value_counts().sort_index()[9:].plot(kind="bar", title="Climate")
climate["years"].value_counts().sort_index()[9:].to_json("../data/vis/numClimate.json", orient="split")
plt.show()
abc = data[data["source"] == "abc"]
abc["years"].value_counts().sort_index().plot(kind="bar", title="ABC")
abc["years"].value_counts().sort_index().to_json("../data/vis/numABC.json", orient="split")
plt.show()
```
